```python
import pandas as pd 

df = pd.read_csv('D:\Data analysis\lego-analysis-master\lego-analysis-master\datasets\lego_sets.csv')
theme = pd.read_csv(r'D:\Data analysis\lego-analysis-master\lego-analysis-master\datasets\parent_themes.csv')
theme.head(50)
```

```python
merged = df.merge(theme, left_on='parent_theme', right_on = 'name')
merged.head()
merged = merged.drop(columns = 'name_y')
merged.head()
```

```python
merged[merged['set_num'].isnull()].shape
```

```python
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
star_wars=licensed[licensed['parent_theme'] == 'Star Wars']
star_wars.head(50)
```

```python
star_wars.shape[0]
```

```python
the_force = int(star_wars.shape[0]/licensed.shape[0]*100)
print(the_force)
```
